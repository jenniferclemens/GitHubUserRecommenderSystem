# Importing JSON File

```python
import urllib, json
from urllib.request import urlopen

#URL of the GitHub users API
url = "https://api.github.com/users"

json_url = urlopen(url)

data = json.loads(json_url.read())

print(data[0]['login'])
data
```

# Converting JSON into Data Frame

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(pd.json_normalize(data), orient='columns')

df
```

```python
df.describe()
```

```python
u = df.loc[0,'following_url']
sub='{/other_user}'
u = u.replace(sub, '') 
print(u)
```

# Expanding the Graph

```python
for i in range(5):
    followingurl=df.loc[i,'following_url']
    sub='{/other_user}'
    followingurl = followingurl.replace(sub, '') 
    followersurl=df.loc[i,'followers_url']
    json_url1 = urlopen(followingurl)
    followingdata = json.loads(json_url1.read())
    json_url2 = urlopen(followersurl)
    followersdata = json.loads(json_url2.read())
    df.append(followingdata)
    df.append(followersdata)
    
```

```python
df.describe()
```

```python
#Removing irrelevant columns

df.drop(["avatar_url", "gravatar_id", "url", "html_url"], axis=1, inplace=True)

df
```
