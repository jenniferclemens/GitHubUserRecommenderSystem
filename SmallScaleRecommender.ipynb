# Creating a Random graph for the list of given nodes

```python
'''https://compucademy.net/generating-random-graphs-in-python/
    Above is the algorithm reference for creating a random graph.'''
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx
from itertools import combinations
from random import random


def ER(n, p):
    V = set([v for v in range(0,n)])
    E = set()
    for combination in combinations(V, 2):
        a = random()
        if a < p:
            E.add(combination)

    g = nx.Graph()
    g.add_nodes_from(V)
    g.add_edges_from(E)

    return g,E


n = 60
p = 0.2
G,E = ER(n, p)
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos)
plt.title("Random Graph Generation")
plt.show()
```

# Sorting the Ranks

```python

tempRanks=ranks

for i in range(len(tempRanks)):
    max_idx = i
    for j in range(i+1, len(tempRanks)):
        if tempRanks[max_idx][1] < tempRanks[j][1]:
            max_idx = j      
    tempRanks[i], tempRanks[max_idx] = tempRanks[max_idx], tempRanks[i]

tempRanks
```
