# Importing JSON File

```python
import urllib, json
from urllib.request import urlopen

#URL of the GitHub users API
url = "https://api.github.com/users"

json_url = urlopen(url)

data = json.loads(json_url.read())

print(data[0]['login'])
data
```

# Converting JSON into Data Frame

```python
import pandas as pd
```

```python
 
df = pd.DataFrame.from_dict(pd.json_normalize(data), orient='columns')

df
```

```python
df.describe()
```

# Expanding the Graph

```python
for i in range(10):
    followingurl=df.loc[i,'following_url']
    sub='{/other_user}'
    followingurl = followingurl.replace(sub, '') 
    followersurl=df.loc[i,'followers_url']
    json_url1 = urlopen(followingurl)
    followingdata = json.loads(json_url1.read())
    json_url2 = urlopen(followersurl)
    followersdata = json.loads(json_url2.read())
    data1 = pd.DataFrame.from_dict(pd.json_normalize(followingdata), orient='columns')
    data2 = pd.DataFrame.from_dict(pd.json_normalize(followersdata), orient='columns')
    df.at[i,'following_url'] = data1
    df.at[i,'followers_url'] = data2
    df=pd.concat([data1,data2], ignore_index=True)
```
